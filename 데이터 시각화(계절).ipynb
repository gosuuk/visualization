```python
# 소스파일 거주지 기준으로 파일 나누기
KO = train.groupby('거주지').get_group('제주도민외')  # 제주도민외 관광객 그룹화 파일
CH = train.groupby('거주지').get_group('중국인')  # 중국인 관광객 그룹화 파일
JJ = train.groupby('거주지').get_group('제주도민') # 제주도민 그룹화 파일
LE = train.groupby('거주지').get_group('법인')  # 법인 그룹화 파일
FG = train.groupby('거주지').get_group('기타외국인') # 기타외국인 그룹화 파일
```

```python
# 날짜 변수에서 연, 월, 일 변수 분리
train['year']=round(train['년월일']/10000)
train['month']=round(((train['년월일']/10000)-train['year'])*100)
train['day']=((((train['년월일']/10000)-train['year'])*100)-train['month'])*100
train
```

```python
#계절구분 
# 3월 ~ 5월 (봄), 6월 ~ 8월(여름), 9월 ~ 11월(가을), 12월 ~ 2월(겨울)로 계절을 나누었습니다.
size=len(train)
season_class={"계절":[]}

for i in range(size):
    if 9<=train["month"][i]<=11:
        season_class["계절"].append("가을")
    elif 3<=train["month"][i]<=5:
        season_class["계절"].append("봄")
    elif 6<=train["month"][i]<=8:
        season_class["계절"].append("여름")
    else : 
        season_class["계절"].append("겨울")
        
season_class=pd.DataFrame(season_class)
train=pd.concat([train,season_class],axis=1)
train
```
